```
import pandas as pd
data = pd.read_csv('/content/drive/My Drive/vedai/data/train_labels.csv')
data = data.drop(["height", "width", "class"], axis=1)
# print (data)


data.rename(
    columns={
        "xmin": "x1",
        "ymin": "y1",
        "xmax": "x2",
        "ymax": "y2"
    },
    inplace=True
)
# Convert the dictionary into DataFrame 
df = pd.DataFrame(data) 

# Declare a list that is to be converted into a column 
listClass = [] 
for x in range(5975):
  listClass.append('car')

# Using 'Address' as the column name 
# and equating it to the list 
df['class'] = listClass 
 
# Observe the result 
# df
for x in range(5975):
  df.at[x,'filename'] = '/content/drive/My Drive/vedai/images/train/' + df.at[x,'filename']

df

df.to_csv(r'/content/drive/My Drive/vedai/data/new-train-anno.csv', index = False)


```

```
import pandas as pd
data = pd.read_csv('/content/drive/My Drive/vedai/data/new-train-anno.csv')
# print (data)


# Convert the dictionary into DataFrame 
df = pd.DataFrame(data) 

for i in range(5975):
  if(df.at[i,'x1'] >= df.at[i,'x2']):
    df.at[i,'x2'] += 1    
  if(df.at[i,'y1'] >= df.at[i,'y2']):
    df.at[i,'y2'] += 1
    

df.to_csv(r'/content/drive/My Drive/vedai/data/train-annotation.csv', index = False)
```
